## Aggregating the order_number grouped by department_id

```python
ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']})
```

Applying the aggregated function to the whole dataframes show that there is orders for each department_id instead of just a selected few. ### Creating  a loyalty flag using transform and loc function

```python
# Creating max_order column
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

```python
# creating loyalty column

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# checking value counts

ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Checking columns

ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

## Determining spending habits between different types of loyal customers

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices':['mean']})
```

For this analysis, it seems like new customers spend on average higher than loyal customers, with regular customers right behind it. I can make an observation and see guess that new customers are given some sort of first time discount and utilizes it by ordering higher priced items, and loyal customers use instacart for more everyday cheaper items that are just considered to much work to get through a store.

## Create a spending flag based on average price

```python
# Creating an average prices column

ords_prods_merge['average_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(15)
```

```python
# connecting the average_price column to a new spending flag column

ords_prods_merge.loc[ords_prods_merge['average_price'] < 10, 'spending_flag'] = 'Low Spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_price'] >=10,'spending_flag']= 'High Spender'
```

```python
# Checking columns

ords_prods_merge[['user_id', 'spending_flag', 'order_number']].head(80)
```

## Creating a frequency flag based on median of days_since_prior_order

```python
# Creating a median_days_since_prior_order column

ords_prods_merge['median_days_since_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head(15)
```

```python
# connecting the median_days_since_prior_order column to a new frequency flag column

ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days_since_prior_order'] > 10) & (ords_prods_merge['median_days_since_prior_order'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] <= 10, 'frequency_flag'] = 'Frequent customer'
```
