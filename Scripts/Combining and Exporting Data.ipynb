## Merging the Dataframes

```python
# the indicator= False is because the df_ords_prods_comb is already including a merge flag
df_orders_products_merged = df_ords_prods_comb.merge(df_prods, on = ['product_id'], indicator = False)
```

## Confirming results with merge flag

```python
df_orders_products_merged['_merge'].value_counts()
```
